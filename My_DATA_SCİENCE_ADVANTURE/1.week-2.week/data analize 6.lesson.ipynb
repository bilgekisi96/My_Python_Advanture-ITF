```python
df6 = pd.DataFrame({'groups': ['A', 'B', 'C', 'A', 'B', 'C','A', 'B', 'C'],
                   'var1': [10,23,33,22,11,99,40,20,15],
                   'var2': [100,253,333,262,111,969, 232,343,123]})
df6
```

```python
df6[["var1","var2"]].transform(lambda x: (x - df6["var1"].min())/(
                                              df6["var1"].max()-df6["var1"].min()))  # make min max scaler
```

```python
df6[["var1","var2"]].apply(lambda x : (x - x.min()) / (x.max() - x.min()))
```

```python
df6[["var1","var2"]].transform(lambda x : (x-x.min())/(x.max()-x.min())) #ikisi ii için
```

```python
df6[["var1","var2"]].apply(lambda x : (x-x.min())/(x.max()-x.min()))
```

```python
df6.groupby("groups").transform(np.mean)
```

```python
df6[["var1"]].transform([np.sqrt,np.exp])
```

```python
df6.groupby("groups").filter(lambda x: x["var2"].mean()>200)
```

```python
#    pivot table
```

```python
titanic.groupby(["sex","pclass"])[["survived"]].mean().unstack() #pivot table
```

```python
titanic.pivot_table(values="survived",index="sex",columns="pclass") #aynısı
```

```python
titanic.groupby(["who","sex","pclass"])[["survived"]].mean().unstack() 
# pclass verisini yukarı almak istiyoruz  unstackleyip yapabiliriz 
#pivottable olarakşuanda survavedin meanleri
```

```python
titanic.pivot_table(values="survived",index=["sex","who"],columns="pclass",aggfunc="count")
#
```

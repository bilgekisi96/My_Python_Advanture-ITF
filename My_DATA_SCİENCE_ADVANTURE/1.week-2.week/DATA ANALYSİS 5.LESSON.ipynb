```python
planet=sns.load_dataset("planets")
```

```python
planet
```

```python
planet.describe()
```

```python
planet.isnull().sum() #null verilerin toplamı
```

```python
planet.dropna() #nul geçen verileri kesipatar
```

```python
planet.method.describe() #içerisinde 10  adert unic değerler varmış
```

```python
planet["method"].value_counts() #uniqe kaç farklı veri toplamını görebiliriz
```

```python
planet["method"].unique() #uniqe değerlere göre groupby yaparız
```

```python
planet.groupby("method").sum()
```

```python
planet.groupby("method")[["distance"]].std()
```

```python
planet.groupby(["method","year"])[["distance","orbital_period"]].std() #içeride bir liste halinde
```

```python
planet.groupby("method")[["orbital_period"]].describe() 
#veri üzerinde efsane işler yapar 
```
